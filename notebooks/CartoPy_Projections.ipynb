### Introduction:
In the world of cartography and GIS, one source of frequent contention is the subject of map projections and the distortions that come with it. To understand this subject let's first begin by looking at the Earth.

![The Earth](https://upload.wikimedia.org/wikipedia/commons/7/7f/Rotating_earth_animated_transparent.gif)

(Animation obtained via [Wikimedia Commons](https://commons.wikimedia.org/wiki/File:Rotating_earth_animated_transparent.gif))

The Earth is a sphere, and with it comes major headaches when attempting to map it out onto a flat surface. If we attempt to flatten the sphere, we find that there is no way to do so without either stretching it, or making cuts to the sphere itself. Therefore, if we attempt to map out the planet, we will find that it is impossible to do so without making the same kind of modifications. This is where map projections come in.

Map projections seek to map out the surface of the planet using a variety of different methods. While there are many different approaches, that involve a variety of different shapes/models, for the purposes of this notebook there are three general projection types that you, the user, should know about. Those being equal-area, and equidistant, and conformal.

### Equal-Area Projections
* Maximize accuracy in terms of the area/size of each region.
* Results in a map that most accurately shows the scales of countries/regions, at the cost of the accuracy of the shape of each.

An example can be seen below with the popular "Mollweide" projection.

```python
#Example 1: Molleweide Projection (Adapted from CartoPy documentation)
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

ax = plt.axes(projection=ccrs.Mollweide())
ax.stock_img()
ax.coastlines()

plt.show()
```

### Equidistant Projection

* Accuracy is focused on the distances between one or more lines from a point, at the cost of scale and shape in other areas.
* Typically used for polar maps/polar-centric maps

An example of this is the azimuthal equidistant projection, which is designed to conserve both distances as well as angles from a central point.

```python
# Example 2: Azimuthal Equidistant Projection (Adapted from CartoPy documentation)
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

ax = plt.axes(projection=ccrs.AzimuthalEquidistant(central_latitude=90))
ax.stock_img()
ax.coastlines()

plt.show()
```

## Conformal Projections
* Shapes and angles of small areas of a map are preserved.
* Results in map that is, generally speaking, proportionately accurate in regards to shape.
* However, this shape accuracy comes at the cost of accuracy in regards to scale 

The difficulties with scale are illustrated by the infamaous Mercator projection seen below.

```python
# Example 3: Mercator Projection (Adapted from CartoPy documentation)
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

ax = plt.axes(projection=ccrs.Mercator())
ax.stock_img()
ax.coastlines()

plt.show()
```

### Which Projection Should I Use?

As is the case with many things, which projection you should use is highly dependant on what you are trying to use the map for. As a typical rule of thumb, the Mercator projection should be avoided as there are usually other map projections that are more appropriate/better designed for a specific use case. Outside of that though, when considering map projections, there are three major questions one should ask.

#### 1. Where is my map of?
If you are mapping for a specific region, it is possible that there is already a preconfigured set of map projections designed to specifically work best for that area. An example of this is the "NAD1983 State Plane Zones" which provides a series of conformal projections designed to work best for any particular US state. For more info [see this article by the US Geological Survey (USGS)](https://www.usgs.gov/faqs/what-state-plane-coordinate-system-can-gps-provide-coordinates-these-values).

#### 2. What am I mapping?
When mapping datasets, especially in GIS tools such as ArcGIS, the dataset you are working with may already be configured for a specific map projection. While it is possible to project these datasets for use in another projection (especially if you are working with multiple datasets or considering creating a map with a different region in mind as previously discussed), if you are just mapping that one particular dataset as it is, it may be best to leave the projection alone.

#### 3. Who is this map for?
As discussed, different projection types work well for different purposes, so it helps to keep in mind what the map ultimately will be used for. For example. if you are making a map that shows how many tornadoes went through a given region for a set period of time, it might be beneficial to stick to a equal area projection. Likewise, if you are building a map with the purpose of navigation, it might be preferable to go with a conformal or equidistant projection.

Regardless of these points however, the fact remains that map projections, as complex as they are, are essential to ensuring that spatial data is presented in an accurate way. While the examples here are meant to showcase differences between different major projection types, they are not all-encompasing and there are many, **many** options for map projections out there. Feel free to try out several and experiment in your next project to see which ones work best for you.
